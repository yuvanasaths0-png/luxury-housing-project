```python
import pandas as pd
chunks=pd.read_csv("/content/Luxury_Housing_Bangalore.csv",chunksize=1000)

output_file="cleaned_data.csv"
first=True

for df in chunks:
    df["Micro_Market"] = df["Micro_Market"].str.strip().str.lower()
    df["Configuration"] = df["Configuration"].str.lower()
    df["Buyer_Type"] = df["Buyer_Type"].str.lower()
    df["Sales_Channel"] = df["Sales_Channel"].str.lower()
    df["Developer_Name"] = df["Developer_Name"].str.lower()
    df["Possession_Status"] = df["Possession_Status"].str.lower()
    df["Transaction_Type"] = df["Transaction_Type"].str.lower()

    df["NRI_Buyer"] = df["NRI_Buyer"].fillna("no")
    df["NRI_Buyer"] = df["NRI_Buyer"].map({"yes":1, "no":0})

    # Convert Amenity_Score to numeric before calculating mean
    df["Amenity_Score"] = pd.to_numeric(df["Amenity_Score"], errors='coerce')
    df["Amenity_Score"] = df["Amenity_Score"].fillna(df["Amenity_Score"].mean())
    df["Buyer_Comments"] = df["Buyer_Comments"].fillna("no comments")

    df["Purchase_Quarter"] = pd.to_datetime(df["Purchase_Quarter"], errors="coerce")

    # Convert columns to numeric before division
    df["Ticket_Price_Cr"] = pd.to_numeric(df["Ticket_Price_Cr"], errors='coerce')
    df["Unit_Size_Sqft"] = pd.to_numeric(df["Unit_Size_Sqft"], errors='coerce')

    df["Price_per_sqft"] = df["Ticket_Price_Cr"]*10000000 / df["Unit_Size_Sqft"]

    df = df[df["Avg_Traffic_Time_Min"] < 120]

    if first:
        df.to_csv(output_file, mode="w", header=True, index=False)
        first = False
    else:
        df.to_csv(output_file, mode="a", header=False, index=False)

    print(df.shape)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
